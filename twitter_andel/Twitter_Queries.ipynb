```python
def count_elastic(es, index, kind, days=1):
    today_date = datetime.today().date()
    today = datetime(today_date.year,today_date.month,today_date.day)
    time_since = today-timedelta(days=days)
    today = today + timedelta(days=1)

    body = {
        #"size":10000,
      "query": {
        "bool": {
            "must": [ 
              {"match_all": {}}
              ],
          #"filter" : {"range": {"date": {"gte": time_since,"lte": today}}}
        }
      }
    }
    
    a = es.count(index = index, doc_type = kind, body = body)
    
    return a
```

```python
def scan_elastic(es, index, kind, days=1):
    today_date = datetime.today().date()
    today = datetime(today_date.year,today_date.month,today_date.day)
    time_since = today-timedelta(days=days)
    today = today + timedelta(days=1)

    body = {
        #"size":10000,
      "query": {
        "bool": {
            "must": [ 
              {"match_all": {}}
              ],
          #"filter" : {"range": {"date": {"gte": time_since,"lte": today}}}
        }
      }
    }
    
    a = scan(client = es, index = index, doc_type = kind, size = 5000, query = body)
    
    return a
```
